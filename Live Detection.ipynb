```python
import numpy as np
import cv2
import tensorflow as tf
import os

emotion_model = tf.keras.models.load_model('Emotion Detection Model.h5')

face_detection_file = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

def restart_kernel():
    cap.release()
    cv2.destroyAllWindows()
    os._exit(0)
   
cap = cv2.VideoCapture(0)

emotion_labels = ['Angry', 'Disgust', 'Fear', 'Happy', 'Sad', 'Surprise', 'Neutral']

try:
    while True:
        _,frame = cap.read()

        grey_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_detection_file.detectMultiScale(grey_frame, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

        for x,y,w,h in faces:
            face = grey_frame[y:y+h,x:x+w]  # Extract the face region in grayscale
            face_resized = cv2.resize(face, (48, 48))  # Resize face to match the input size of the emotion detection model
            face_resized = np.expand_dims(face_resized, axis=-1)  # Add a channel dimension for compatibility
            face_resized = np.expand_dims(face_resized, axis=0)  # Add a batch dimension


            prediction = emotion_model.predict(face_resized) 

            # Get the predicted emotion label
            predicted_emotion_index = np.argmax(prediction)
            predicted_emotion = emotion_labels[predicted_emotion_index]

            # Draw rectangle around the face and label the emotion
            cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
            cv2.putText(frame, predicted_emotion, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

        cv2.imshow("window", frame)

        if cv2.waitKey(1)==ord('q'):
            restart_kernel()

except KeyboardInterrupt:
    restart_kernel()
```
